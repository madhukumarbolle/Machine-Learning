```python
# Visualization the image of first 5 digital in X - also print the class label in

# for i in range(0,5):
#   img = X.iloc[i].values.reshape(28,28)
#   label = y.iloc[i]
#   plt.imshow(img,cmap = "gray")
#   plt.title(f"label: {label}")
#   plt.show()




plt.figure(figsize=(12, 4))  # Set figure size

for i in range(5):
    img = X.iloc[i].values.reshape(28, 28)  # Reshape to 28x28 image
    label = y.iloc[i]  # Get label

    plt.subplot(1, 5, i + 1)  # 1 row, 5 columns, position i+1
    plt.imshow(img, cmap="gray")  # Display grayscale image
    plt.title(f"Label: {label}")  # Set title
    plt.axis("off")  # Hide axis for better visualization

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
# 11
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.25,random_state = 42)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=315,metric = "euclidean")
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred= knn.predict(X_test)
```

```python
# from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
